> This notebook contains code to count the number of detections in a single trajectory

## Python Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 150
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif"
})
plt.rcParams['font.size'] = '18'
```

## Load data

```python
fr = pd.read_csv('../data/2022-09-06-12-55-08/front_right.csv')
```

```python
fl = pd.read_csv('../data/2022-09-06-12-55-08/front_left.csv')
```

```python
rr = pd.read_csv('../data/2022-09-06-12-55-08/rear_right.csv')
```

```python
rl = pd.read_csv('../data/2022-09-06-12-55-08/rear_left.csv')
```

## Plotting data

```python
fr = fr.groupby(['Time']).count()
```

```python
fl = fl.groupby(['Time']).count()
```

```python
rr = rr.groupby(['Time']).count()
```

```python
rl = rl.groupby(['Time']).count()
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(4, 1, figsize=[12.8, 6.8], sharex=True)
# colors = [plt.cm.winter(i) for i in np.linspace(0.2, 0.95, 4)] #
colors = ['darkgrey','darkgrey','darkgrey','darkgrey']
# ax.set_prop_cycle('color', [plt.cm.viridis(i) for i in np.linspace(0, 1, 4)])
ax1.fill_between((fr.index.to_numpy()-fr.index.to_numpy()[0])/60, fr.speed.to_numpy(), color=colors.pop(0), label='front right')
ax2.fill_between((fl.index.to_numpy()-fl.index.to_numpy()[0])/60, fl.speed.to_numpy(), color=colors.pop(0), label='front left')
ax3.fill_between((rr.index.to_numpy()-rr.index.to_numpy()[0])/60, rr.speed.to_numpy(), color=colors.pop(0), label='rear right')
ax4.fill_between((rl.index.to_numpy()-rl.index.to_numpy()[0])/60, rl.speed.to_numpy(), color=colors.pop(0), label='rear left')
ax4.set_xlabel('Time [min]')
ax1.set_ylabel('fr')
ax2.set_ylabel('fl')
ax3.set_ylabel('rr')
ax4.set_ylabel('rl')
# ax1.legend(loc='upper left', fancybox=False)
# ax2.legend(loc='upper left', fancybox=False)
# ax3.legend(loc='upper left', fancybox=False)
# ax4.legend(loc='upper left', fancybox=False)
# ax1.set_title('front right')
# ax2.set_title('front left')
# ax3.set_title('rear right')
# ax4.set_title('rear left')
ax1.grid(linestyle=':', color='dimgray',linewidth=1, axis='both')
ax2.grid(linestyle=':', color='dimgray',linewidth=1, axis='both')
ax3.grid(linestyle=':', color='dimgray',linewidth=1, axis='both')
ax4.grid(linestyle=':', color='dimgray',linewidth=1, axis='both')
ax1.set_xlim([0, np.max(fr.index.to_numpy()-fr.index.to_numpy()[0])/60])
ax1.set_ylim(bottom=0)
ax2.set_ylim(bottom=0)
ax3.set_ylim(bottom=0)
ax4.set_ylim(bottom=0)
plt.tight_layout()
plt.show()
```

```python
fig.savefig('../images/2022-09-06-12-55-08/detections.png', dpi=300)
```
