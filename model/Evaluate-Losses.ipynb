```python
import os
import pathlib
import time
import random
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 150
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif"
})
plt.rcParams['font.size'] = '18'
```

```python
fr = torch.load('../checkpoints/2023-04-10-06-58-41-RadarNet-Short-V0/losses.pt')
fl = torch.load('../checkpoints/2023-04-24-17-44-19-RadarNet-Short-V0-FL/losses.pt')
rr = torch.load('../checkpoints/2023-04-30-18-06-16-RadarNet-Short-V1/losses.pt')
rl = torch.load('../checkpoints/2023-05-01-22-23-39-RadarNet-Short-V1-RL/losses.pt')
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(12,4), sharey=True)
ax1.loglog(fr['train_losses'],'b',label='Train')
ax1.loglog(fr['test_losses'],'r',label='Test')
ax2.loglog(fl['train_losses'],'b',label='Train')
ax2.loglog(fl['test_losses'],'r',label='Test')
ax1.grid(linestyle=':')
ax2.grid(linestyle=':')
ax1.legend(fancybox=False)
ax2.legend(fancybox=False)
ax1.set_ylabel('$m^2/s^2$')
ax1.set_xlabel('Epochs')
ax2.set_xlabel('Epochs')
ax1.set_xlim(left=1,right=1000)
ax2.set_xlim(left=1,right=100)
# ax1.set_ylim(bottom=0)
plt.tight_layout()
plt.show()
```

```python
fig.savefig('../output/front-radars-learning.png', dpi=300);
```
